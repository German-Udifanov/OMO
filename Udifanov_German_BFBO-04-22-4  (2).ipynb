```python
#Пример 10
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import numpy as np 
import seaborn as sns

iris = sns.load_dataset('iris')
iris   

X_train, X_test,y_train, y_test = train_test_split(
    iris.iloc[:, :-1],#все колонки кроме последней 
    iris.iloc[:,-1],#последнюю в целевую переменную
    test_size = 0.20 
)

X_train.shape, X_test.shape,y_train.shape, y_test.shape

X_train.head()
y_train.head()

#обучим метод трёх ближайших соседей
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)

#Получим предсказаниемодели 
y_pred = model.predict(X_test)
y_pred 

#покажем на графике что отображает полученное число
#красным цветом отображаются точки, для которых классификация сработала неправильно
plt.figure(figsize=(10,7))
sns.scatterplot(x='petal_width',y='petal_length', data = iris, hue = 'species', s=70)
plt.xlabel('Длинна лепестков, см')
plt.ylabel('Длинна лепестков, см')
plt.legend(loc=2)
plt.grid()

#Переберём все объекты из теста
for i in range(len(y_test)):
    #Если пркедсказание не правильное 
    if np.array(y_test)[i] != y_pred[i]:
        #то подсвечивать точку красным
        plt.scatter(X_test.iloc[i,3],X_test.iloc[i,2],color='red', s=150)
#качество моделеи (доля правильно классиффицированных точек)
from sklearn.metrics import accuracy_score
print(f'accuracy:{accuracy_score(y_test,y_pred) :.3} ')
```
